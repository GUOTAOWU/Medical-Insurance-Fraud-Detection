# 样本类别不平衡

```python
import os
import pandas as pd
import joblib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from collections import Counter
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
from imblearn.combine import SMOTETomek

# 设置中文显示和负号显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 常量定义
INPUT_FILE = '最终数据.xlsx'
OUTPUT_DIR = 'processed_data'
TEST_SIZE = 0.2
RANDOM_STATE = 42

# 1. 读取数据
data = pd.read_excel(INPUT_FILE)
X = data.drop('fraud', axis=1)  # 特征矩阵（所有列，除了 fraud）
y = data['fraud']  # 目标变量（fraud 列）

# 2. 检查目标变量分布
print(f"原始目标变量类别分布: {Counter(y)}")

# 3. 分割数据集
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=TEST_SIZE,
    stratify=y,  # 保持类别比例
    random_state=RANDOM_STATE
)

# 4. 应用SMOTE+Tomek组合采样
smote_tomek = SMOTETomek(
    sampling_strategy='auto',  # 自动平衡类别
    random_state=RANDOM_STATE,
    smote=SMOTE(k_neighbors=5),  # SMOTE过采样
    tomek=TomekLinks()  # Tomek欠采样
)
X_resampled, y_resampled = smote_tomek.fit_resample(X_train, y_train)
print(f"处理后训练集类别分布: {Counter(y_resampled)}")

# 5. 可视化数据分布对比
plt.figure(figsize=(12, 4))

# 原始训练集分布
plt.subplot(121)
plt.pie(Counter(y_train).values(), labels=Counter(y_train).keys(), autopct='%1.1f%%')
plt.title('原始训练集分布')

# 采样后训练集分布
plt.subplot(122)
plt.pie(Counter(y_resampled).values(), labels=Counter(y_resampled).keys(), autopct='%1.1f%%')
plt.title('SMOTE+Tomek处理后分布')

plt.tight_layout()
plt.show()

# 6. 保存处理后的数据
os.makedirs(OUTPUT_DIR, exist_ok=True)

# 保存训练集和测试集
train_df = pd.concat([X_resampled, pd.Series(y_resampled, name='fraud')], axis=1)
test_df = pd.concat([X_test, pd.Series(y_test, name='fraud')], axis=1)
train_df.to_csv(f'{OUTPUT_DIR}/resampled_train.csv', index=False)
test_df.to_csv(f'{OUTPUT_DIR}/original_test.csv', index=False)

# 保存采样器
joblib.dump(smote_tomek, f'{OUTPUT_DIR}/sampler.pkl')

print(f"\n数据已保存到 {OUTPUT_DIR} 目录:")
print(f"  - resampled_train.csv : 过采样后的训练集")
print(f"  - original_test.csv   : 原始测试集")
print(f"  - sampler.pkl         : 采样器对象")
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report, f1_score, roc_auc_score
import joblib
from sklearn.utils import parallel_backend
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")

# 1. 定义模型评估函数
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print(classification_report(y_test, y_pred))
    print(f"Macro F1-score: {f1_score(y_test, y_pred, average='macro'):.4f}")
    try:
        y_proba = model.predict_proba(X_test)
        print(f"ROC-AUC (OvO): {roc_auc_score(y_test, y_proba, multi_class='ovo'):.4f}")
    except:
        pass
    return f1_score(y_test, y_pred, average='macro')

# 2. 初始化模型
models = {
    "RandomForest": RandomForestClassifier(n_estimators=200, class_weight='balanced', random_state=42),
    "XGBoost": XGBClassifier(n_estimators=300, learning_rate=0.1, eval_metric='mlogloss'),
}

# 3. 训练并评估传统模型
best_model = None
best_score = -1  # 初始化为-1，确保能捕获所有可能的分值
best_test_score = -1  # 存储最佳模型的测试集分数
model_scores = {}  # 用于存储所有模型得分
model_test_scores = {}  # 用于存储所有模型的测试集分数

for name, model in models.items():
    print(f"\n=== {name} ===")
    model.fit(X_resampled, y_resampled)

    # 交叉验证
    cv_scores = cross_val_score(model, X_resampled, y_resampled, cv=5, scoring='f1_macro', n_jobs=-1)
    mean_cv_score = np.mean(cv_scores)
    print(f"CV F1-score: {mean_cv_score:.4f} (±{np.std(cv_scores):.4f})")

    # 测试集评估
    current_test_score = evaluate_model(model, X_test, y_test)
    model_scores[name] = mean_cv_score  # 存储交叉验证平均分
    model_test_scores[name] = current_test_score  # 存储测试集分数

    # 保存最佳模型
    if (mean_cv_score > best_score) or (abs(mean_cv_score - best_score) < 1e-6 and current_test_score > best_test_score):
        best_score = mean_cv_score
        best_test_score = current_test_score
        best_model = name
        joblib.dump(model, 'best_traditional_model.pkl')
        print(f"更新最佳模型: {name} (CV: {mean_cv_score:.4f}, Test: {current_test_score:.4f})")

# 4. 训练并评估SVM模型
with parallel_backend('threading', n_jobs=-1):  # 使用多线程加速
    print("\n=== SVM ===")
    svm_model = make_pipeline(
        StandardScaler(),  # 数据标准化
        SVC(kernel='linear',C=0.1, class_weight='balanced', probability=True, verbose=True, max_iter=2000, tol=1e-3, random_state=42, cache_size=1000)
    )
    svm_model.fit(X_resampled, y_resampled)
    svm_cv_scores = cross_val_score(svm_model, X_resampled, y_resampled, cv=5, scoring='f1_macro')
    mean_svm_score = np.mean(svm_cv_scores)
    print(f"CV F1-score: {mean_svm_score:.4f} (±{np.std(svm_cv_scores):.4f})")
    svm_test_score = evaluate_model(svm_model, X_test, y_test)
    models["SVM"] = svm_model
    model_scores["SVM"] = mean_svm_score
    model_test_scores["SVM"] = svm_test_score

    # 检查SVM是否成为最佳模型
    if (mean_svm_score > best_score) or (abs(mean_svm_score - best_score) < 1e-6 and svm_test_score > best_test_score):
        best_score = mean_svm_score
        best_test_score = svm_test_score
        best_model = "SVM"
        joblib.dump(svm_model, 'best_traditional_model.pkl')
        print(f"更新最佳模型: SVM (CV: {mean_svm_score:.4f}, Test: {svm_test_score:.4f})")

# 5. 训练并评估神经网络模型（MLP）
print("\n=== MLPClassifier ===")
mlp = MLPClassifier(hidden_layer_sizes=(128, 64), activation='relu', early_stopping=True, validation_fraction=0.2, random_state=42)
mlp.fit(X_resampled, y_resampled)
mlp_cv_scores = cross_val_score(mlp, X_resampled, y_resampled, cv=5, scoring='f1_macro')
mean_mlp_score = np.mean(mlp_cv_scores)
print(f"CV F1-score: {mean_mlp_score:.4f} (±{np.std(mlp_cv_scores):.4f})")
mlp_test_score = evaluate_model(mlp, X_test, y_test)
joblib.dump(mlp, 'mlp_model.pkl')
model_scores["MLP"] = mean_mlp_score
model_test_scores["MLP"] = mlp_test_score

# 检查MLP是否成为最佳模型
if (mean_mlp_score > best_score) or (abs(mean_mlp_score - best_score) < 1e-6 and mlp_test_score > best_test_score):
    best_score = mean_mlp_score
    best_test_score = mlp_test_score
    best_model = "MLP"
    joblib.dump(mlp, 'best_traditional_model.pkl')
    print(f"更新最佳模型: MLP (CV: {mean_mlp_score:.4f}, Test: {mlp_test_score:.4f})")

# 6. 模型比较与选择
print("\n=== 模型性能比较 ===")
for name, score in sorted(model_scores.items(), key=lambda x: (x[1], model_test_scores[x[0]]), reverse=True):
    print(f"{name}: CV F1-score = {score:.4f}, Test F1-score = {model_test_scores[name]:.4f}")

print(f"\n最佳模型: {best_model} (CV F1: {best_score:.4f}, Test F1: {best_test_score:.4f})")
```

```python
#验证曲线调参
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve, validation_curve
from xgboost import XGBClassifier
from sklearn.metrics import f1_score, classification_report
import joblib
from time import time
from xgboost import plot_importance
import warnings
from sklearn.exceptions import FitFailedWarning

# 忽略 FitFailedWarning 警告
warnings.filterwarnings("ignore", category=FitFailedWarning)

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 加载或初始化模型
try:
    model = joblib.load('best_traditional_model.pkl')
    print("成功加载已保存的模型")
except FileNotFoundError:
    print("未找到保存的模型，初始化新模型")
    model = XGBClassifier(n_estimators=300, learning_rate=0.1, random_state=42)

# 定义学习曲线绘制函数
def plot_learning_curve(estimator, title, X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 5), metric='f1_macro'):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, scoring=metric, train_sizes=train_sizes, n_jobs=-1, random_state=42
    )
    train_mean, train_std = np.mean(train_scores, axis=1), np.std(train_scores, axis=1)
    test_mean, test_std = np.mean(test_scores, axis=1), np.std(test_scores, axis=1)

    plt.figure(figsize=(12, 6))
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_mean, 'o-', color="r", label="训练集得分")
    plt.plot(train_sizes, test_mean, 'o-', color="g", label="验证集得分")

    plt.xlabel("训练样本数")
    plt.ylabel(f"{metric}分数")
    plt.title(title)
    plt.legend(loc="best")
    plt.grid(True)
    plt.show()

# 初始模型评估
print("\n=== 初始模型性能 ===")
plot_learning_curve(model, "初始模型学习曲线", X_resampled, y_resampled)

# 定义验证曲线绘制函数
def plot_validation_curve(estimator, title, X, y, param_name, param_range, cv=5, metric='f1_macro'):
    train_scores, test_scores = validation_curve(
        estimator, X, y, param_name=param_name, param_range=param_range, cv=cv, scoring=metric, n_jobs=-1
    )
    train_mean, train_std = np.mean(train_scores, axis=1), np.std(train_scores, axis=1)
    test_mean, test_std = np.mean(test_scores, axis=1), np.std(test_scores, axis=1)

    plt.figure(figsize=(10, 6))
    plt.plot(param_range, train_mean, 'o-', color="r", label="训练集")
    plt.plot(param_range, test_mean, 'o-', color="g", label="验证集")
    plt.fill_between(param_range, train_mean - train_std, train_mean + train_std, alpha=0.1, color="r")
    plt.fill_between(param_range, test_mean - test_std, test_mean + test_std, alpha=0.1, color="g")

    plt.xlabel(param_name)
    plt.ylabel(f"{metric}分数")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

    return param_range[np.argmax(test_mean)]

# 调优 n_estimators
print("\n=== 调优 n_estimators ===")
best_n = plot_validation_curve(
    XGBClassifier(learning_rate=0.1, random_state=42),
    "n_estimators 验证曲线", X_resampled, y_resampled,
    param_name="n_estimators", param_range=[100, 200, 300, 400, 500]
)
print(f"最佳 n_estimators: {best_n}")

# 调优 max_depth
print("\n=== 调优 max_depth ===")
best_depth = plot_validation_curve(
    XGBClassifier(n_estimators=best_n, learning_rate=0.1, random_state=42),
    "max_depth 验证曲线", X_resampled, y_resampled,
    param_name="max_depth", param_range=[3, 5, 7, 9, 11]
)
print(f"最佳 max_depth: {best_depth}")

# 调优 learning_rate
print("\n=== 调优 learning_rate ===")
best_lr = plot_validation_curve(
    XGBClassifier(n_estimators=best_n, max_depth=best_depth, random_state=42),
    "learning_rate 验证曲线", X_resampled, y_resampled,
    param_name="learning_rate", param_range=[0.01, 0.05, 0.1, 0.2, 0.3]
)
print(f"最佳 learning_rate: {best_lr}")

# 训练最终模型
print("\n=== 训练最终模型 ===")
final_model = XGBClassifier(
    n_estimators=best_n,
    max_depth=best_depth,
    learning_rate=best_lr,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='mlogloss',
    random_state=42
)

start_time = time()
final_model.fit(X_resampled, y_resampled)
print(f"训练时间: {time() - start_time:.2f} 秒")

# 评估最终模型
plot_learning_curve(final_model, "最终模型学习曲线", X_resampled, y_resampled)

# 保存调优后的模型
joblib.dump(final_model, 'optimized_xgboost_model.pkl')
print("已保存调优后的模型")

# 测试集评估
y_pred = final_model.predict(X_test)
print("\n=== 测试集性能 ===")
print(classification_report(y_test, y_pred))
print(f"Macro F1-score: {f1_score(y_test, y_pred, average='macro'):.4f}")

# 特征重要性可视化
plot_importance(final_model, max_num_features=20)
plt.show()
```

```python
#网格搜索调参
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, learning_curve, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neural_network import MLPClassifier
from sklearn.inspection import partial_dependence
import xgboost as xgb
import joblib
from time import time
```

```python
base_params = {
    'objective': 'multi:softmax',  # 多分类任务
    'num_class': 5,               # 类别数
    'eval_metric': 'mlogloss',
    'n_estimators': 500,
    'random_state': 42,
    'n_jobs': -1
}

# 3. 分阶段网格搜索
def xgb_grid_search(phase_name, param_grid, model=None):
    print(f"\n=== Phase {phase_name} Grid Search ===")
    start_time = time()
    
    if model is None:
        model = xgb.XGBClassifier(**base_params)
    
    grid = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        cv=5,
        scoring='accuracy',
        verbose=2,
        n_jobs=-1
    )
    grid.fit(X_resampled, y_resampled, 
             eval_set=[(X_test, y_test)],
             verbose=False)
    
    print(f"Best params: {grid.best_params_}")
    print(f"Best score: {grid.best_score_:.4f}")
    print(f"Time cost: {time()-start_time:.1f}s")
    
    return grid.best_estimator_, grid.cv_results_

# 第一阶段：核心参数
phase1_params = {
    'learning_rate': [0.3],
    'max_depth': [7],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0]
}
best_phase1, cv_results1 = xgb_grid_search("1 - Core Params", phase1_params)

# 第二阶段：正则化参数
phase2_params = {
    'gamma': [0, 0.1, 0.5],
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [0.1, 1, 10],
    'min_child_weight': [1, 3, 5]
}
best_phase2, cv_results2 = xgb_grid_search("2 - Regularization", phase2_params, best_phase1)

```

```python
# 4. 可视化分析
plt.figure()

# 学习曲线

train_sizes, train_scores, val_scores = learning_curve(
    best_phase2, X_resampled, y_resampled, cv=5, scoring='accuracy')
plt.plot(train_sizes, np.mean(train_scores, axis=1), 'o-', label='Train')
plt.plot(train_sizes, np.mean(val_scores, axis=1), 'o-', label='Validation')
plt.title('Learning Curve')
plt.xlabel('Training examples')
plt.ylabel('Accuracy')
plt.legend()

# 特征重要性
xgb.plot_importance(best_phase2)
plt.title('Feature Importance')
```

```python
# 混淆矩阵
y_pred = best_phase2.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.title('Confusion Matrix')
```

```python
# 参数热力图（示例：learning_rate vs max_depth）
heatmap_data = pd.pivot_table(
    pd.DataFrame(cv_results1),
    values='mean_test_score',
    index='param_learning_rate',
    columns='param_max_depth'
)
sns.heatmap(heatmap_data, annot=True, fmt=".3f")
plt.title('Phase1 Grid Search Results')
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# 5. MLP对比验证
print("\n=== MLP Comparison ===")
mlp = MLPClassifier(
    hidden_layer_sizes=(100, 50),
    activation='relu',
    solver='adam',
    early_stopping=True,
    validation_fraction=0.2,
    verbose=True,
    random_state=42
)
mlp.fit(X_resampled, y_resampled)

# MLP损失曲线
plt.figure(figsize=(8, 4))
plt.plot(mlp.loss_curve_)
plt.title('MLP Training Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.savefig('mlp_loss_curve.png')
plt.show()
```

```python
# 6. 模型评估
def evaluate_model(name, model, X_test, y_test):
    start_time = time()
    y_pred = model.predict(X_test)
    
    print(f"\n{name} Evaluation:")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(f"Classification Report:\n{classification_report(y_test, y_pred)}")
    print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}")
    print(f"Prediction Time: {(time()-start_time)*1000:.2f}ms")
    
    return y_pred

xgb_pred = evaluate_model("XGBoost", best_phase2, X_test, y_test)
mlp_pred = evaluate_model("MLP", mlp, X_test, y_test)
```
